# Why study the distributions ?
In this code example, we will be looking at the various statistical distributions, showing the under lying relation among the data points. This relation is important to decide what kind of model is used for what kind of dataset.

Distributions that we will be discussing here: 
- Normal distribution 
- Binomial distribution 
- Gamma distribution 
- Uniform distribution 
- Poisson distribution 
- Tweedie distribution 

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings('ignore')
```

# Normal distribution 


```python
# Some random initial parameters 
mean = 1
standard_deviation = 2
sample_points = 10000

norm_distrib = np.random.normal(mean, standard_deviation, sample_points)

fig = sns.distplot(pd.DataFrame(norm_distrib)).set_title(
    'Normal distribution with mean = %f \
    and standard deviation = %f'%(mean, standard_deviation))
```

```python
# If we set the mean = 0 and standard deviation to 1, then this
# distribution is called standard normal distribution.

mean = 0; standard_deviation = 1;

norm_distrib = np.random.normal(mean, standard_deviation, sample_points)

fig = sns.distplot(pd.DataFrame(norm_distrib)).set_title(
    'Normal distribution with mean = %f \
    and standard deviation = %f'%(mean, standard_deviation))
```

```python
# Verifying the values of the mean and variance from the dataset 
# that we created using the numpy function

print('Mean = ', np.mean(norm_distrib)) 
# this value should be approximately 0; approximately since the 
# precision depends on the number of samples you have taken while 
# sampling using the numpy function 

# Variance = (standard_deviation)^2
print('Variance = ', np.var(norm_distrib))
# similarly, this value should be close to 1
```

```python
# Let's observe the curve for different number of samples 

for samples in np.arange(1, 20, 5):
    norm_distrib = np.random.normal(mean, standard_deviation, samples*100)
    plt.figure(samples)
    fig = sns.distplot(pd.DataFrame(norm_distrib)).set_title(
    'Normal distribution with mean = %f \
    and standard deviation = %f and %d sample points'%(mean, standard_deviation, samples*1000))
```

# Uniform continuous distribution
In this type of distribution, probability associated with each point in
a certain fixed interval is same. 

```python
low = 1
high = 5
sample_points = 10000

uniform_distrib = np.random.uniform(low, high, sample_points)

fig = sns.distplot(pd.DataFrame(uniform_distrib)).set_title(
    'Uniform distribution in the range %d - %d'%(low, high))
```

```python
# Calculating the values of the mean and variance

print('Mean = ', np.mean(uniform_distrib)) 
# this value should be close to ((low + high)/2); approximately since the 
# precision depends on the number of samples you have taken while 
# sampling using the numpy function 

# Variance = (standard_deviation)^2
print('Variance = ', np.var(uniform_distrib))
# this value should be close to (((high - low)^2)/12)
```

# Binomial distribution

A discrete probability distribution, giving the sum of outcomes of 'n' bernoulli trials. 

```python
num_trials = 100
prob_success = 0.5
sample_points = 1000

binomial_distrib = np.random.binomial(num_trials, prob_success, sample_points)

fig = sns.distplot(pd.DataFrame(binomial_distrib)).set_title(
    'Binomial distribution for %d trials and \
    probability of success = %f'%(num_trials, prob_success))
```

```python
# Calculating the values of the mean and variance

print('Mean = ', np.mean(binomial_distrib)) 

# Variance = (standard_deviation)^2
print('Variance = ', np.var(binomial_distrib))

# Variance￼ is ￼(1-prob_success) times the mean￼ and will always be less than  mean.
```

# Poisson distribution
A discrete probability distribution, giving the number of times an event occured in a fixed time interval. 
Example: Number of cars passing by a certain traffic junction in a minute. 

This distribution is an approximation of the binomial distribution, where the number of trials are sufficiently large and the probability of success is significantly small. 


```python
# lambda = num_trials*prob_success
lam = 3  
sample_points = 10000

poisson_distrib = np.random.poisson(lam, sample_points)

fig = sns.distplot(pd.DataFrame(poisson_distrib)).set_title(
    'Poisson distribution for lambda = %f and \
    size = %d'%(lam, sample_points))
```

```python
# Calculating the values of the mean and variance

print('Mean = ', np.mean(poisson_distrib)) 

# Variance = (standard_deviation)^2
print('Variance = ', np.var(poisson_distrib))

# Variance and mean are almost same and are very close 
# to lambda in poisson distribution
```

# Gamma distribution

A distribution used for continuous variables. Employs two special paramters: shape and scale. Shape defines how the distribution will look, while scale parameter will define where bulk of the observations lie in the distribution.

Mean = shape * scale

Variance = shape * (scale)^2

```python
shape = 1
scale = 1.5
sample_points = 100000

gamma_distrib = np.random.gamma(shape, scale, sample_points)

fig = sns.distplot(pd.DataFrame(gamma_distrib)).set_title(
    'Gamma distribution for shape = %d and scale = %f'%(shape, scale))
```

```python
# Calculating the values of the mean and variance
print('Mean = ', np.mean(gamma_distrib)) 

# Variance = (standard_deviation)^2
print('Variance = ', np.var(gamma_distrib))
```

```python
# Let's see how the graph looks for other values of shape and scale 
for shape, scale in zip(np.arange(1, 3, 1), np.arange(1, 5, 2)):
    gamma_distrib = np.random.gamma(shape, scale, sample_points)
    plt.figure(shape)
    fig = sns.distplot(pd.DataFrame(gamma_distrib)).set_title(
    'Gamma distribution for shape = %d and scale = %f'%(shape, scale))
```
